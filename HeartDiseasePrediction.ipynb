```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dat = load_breast_cancer()
```

```python
df = pd.DataFrame(data=dat.data, columns=dat.feature_names)
```

```python
df.shape
```

```python
df
```

```python
y = dat['target']
```

```python
type(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df,y)
```

```python
print("X Train data : ",X_train.shape)
print("\ny Train data : ",y_train.shape)
print("\nX Test data : ",X_test.shape)
print("\ny Test data : ",y_test.shape)
```

## Using Naive Baye's Algorithm 

```python
from sklearn.naive_bayes import GaussianNB

mod = GaussianNB()
mod_NB = mod.fit(X_train,y_train)
```

```python
y_pred_NB = mod_NB.predict(X_test)
```

```python
#Lets evaluate the model 
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

print(confusion_matrix(y_test,y_pred_NB))
```

```python
accuracy_score(y_test,y_pred_NB)
```

```python
print(classification_report(y_test,y_pred_NB))
```

## Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(solver="saga",max_iter=5000)

params = {'C':np.logspace(-5,5,20)}
model_LR = GridSearchCV(model,param_grid=params,cv=10,scoring="accuracy")
```

```python
model_LR.fit(X_train,y_train)
```

```python
y_pred_LR = model_LR.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_LR)
```

```python
print(classification_report(y_test,y_pred_LR))
```

## Comparision of both models 

```python
import matplotlib.pyplot as plt 

model_names = ["Naive Baye's","Logistic Regression"]

score_NB = accuracy_score(y_test,y_pred_NB)*100
score_LR = accuracy_score(y_test,y_pred_LR)*100

model_scores = [score_NB,score_LR]


plt.figure(figsize=(5, 5))
bars = plt.bar(model_names,model_scores, color=["Blue","Green"])
plt.bar_label(bars, labels= [score_NB,score_LR])
plt.title("Model Performance")
plt.show
```
